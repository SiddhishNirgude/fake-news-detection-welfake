## EDA Key Findings

- **Dataset size**: 72,134 articles (35,028 Fake / 37,106 Real)

- **Vocabulary**: 188,452 unique words in text; both classes share similar top keywords
- **Short articles**: 1,847 articles under 20 words — may need filtering
- **Non-ASCII**: 35,088 articles contain non-ASCII characters — consider normalization
